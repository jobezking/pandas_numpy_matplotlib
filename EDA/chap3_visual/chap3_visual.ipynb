```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Load Titanic dataset directly from seaborn
titanic = sns.load_dataset('titanic')
```

```python
mean_age = titanic['age'].mean()
median_age = titanic['age'].median()
mode_age = titanic['age'].mode()[0]
print(mean_age, median_age, mode_age)
```

```python
var_age = titanic['age'].var() #Variance indicates how far values are spread around the mean by averaging squared deviations.
std_age = titanic['age'].std()
print(var_age, std_age)
```

```python
plt.hist(titanic['age'].dropna(),bins=20)
plt.title("Age Distribution of Titanic Passengers")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.kdeplot(titanic['age'].dropna(),fill=True)
plt.title("Age Density Plot - Titanic Distribution")
plt.xlabel("Age")
plt.show()
```
