```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
titanic = pd.read_csv('titanic_cleaned.csv')
```

```python
sns.countplot(x='sex', hue='survived', data=titanic)
plt.title("Survival Distribution by Sex")
plt.show()
```

```python
sns.countplot(x='pclass', hue='survived', data=titanic) 
plt.title("Survival Distribution by Passenger Class") 
plt.show()
```

```python
sns.catplot(x='pclass', hue='sex', col='survived', kind='count', data=titanic)
plt.show()
```

```python
sns.histplot(data=titanic, x='age', hue='survived', multiple='stack', bins=20) 
plt.title("Age Distribution by Survival") 
```

```python
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title("Age Distribution Across Classes")
plt.show()
```

```python
sns.violinplot(x='survived',y='fare',data=titanic)
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
#Create combined features
titanic['FamilySize'] = titanic['sibsp'] + titanic['parch'] + 1
```

```python
sns.barplot(x='FamilySize', y='survived', data=titanic)
plt.title("Survival Rate by Family Size") 
plt.show()
```

```python
titanic['IsAlone'] = 1  #sets every column to traveling alone because values will be either 1 or 0
titanic.loc[titanic['FamilySize'] > 1, 'IsAlone'] = 0 # if family size is greater than 1, not alone. set field to 0
sns.barplot(x='IsAlone', y='survived', data=titanic)
plt.title("Survival Rate: Alone vs With Family")
plt.show()
```
