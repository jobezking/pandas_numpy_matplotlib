```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load Titanic dataset directly from seaborn
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#Univariate Analysis
```

```python
sns.histplot(df['age'],kde=True)  #kde is kernel density estimate curve to show probability distribution
plt.title('Age Distribution of Passengers')
plt.show()
```

```python
#sex distribution
sns.countplot(x='sex',data=df)
plt.title('Passenger Count by Gender')
plt.show()
```

```python
#Bivariate Analysis
```

```python
#plot survival vs gender
sns.barplot(x='sex', y='survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='pclass',y='survived',data=df)
plt.title('Survival Rate by class')
plt.show()
```

```python
#Multivariate Analysis
```

```python
sns.catplot(x='pclass',y='survived',hue='sex',data=df, kind='bar')
plt.title('Titanic Survival Rate by Class and Sex')
plt.show()
```

```python
#Check for missing data
sns.heatmap(df.isnull(),cbar=False)
plt.title('Missing Values in Titanic Dataset')
plt.show()
```
