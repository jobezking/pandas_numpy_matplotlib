```python
df = pd.read_csv("healthcare-dataset-stroke-data.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df['bmi'] = df['bmi'].fillna(df['bmi'].median())
```

```python
df.isnull().sum()
```

```python
scaler = StandardScaler()  #scaling numeric features
df[['age','avg_glucose_level','bmi']] = scaler.fit_transform(df[['age','avg_glucose_level','bmi']])
```

```python
df['glucose_log'] = np.log(df['avg_glucose_level'] + 1) #handling skewness
```

```python
df = pd.get_dummies(df, drop_first=True)
```

```python
df['age_risk'] = df['age'].apply(lambda x: 1 if x > 60 else 0)
df['bmi_obese'] = df['bmi'].apply(lambda x: 1 if x > 30 else 0)
```

```python
df['hypertension_glucose'] = df['hypertension'] * df['glucose_log']
```
