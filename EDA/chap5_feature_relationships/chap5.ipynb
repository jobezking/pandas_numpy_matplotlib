```python
from sklearn.datasets import fetch_california_housing
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = fetch_california_housing(as_frame=True, data_home='./temp_data')
df = data.frame
df.head()
```

```python
df.corr()['MedHouseVal'].sort_values(ascending=False) #calcuate correlation with target variable MedHouseVal
```

```python
sns.scatterplot(data=df,x='MedInc',y='MedHouseVal')
plt.title("House Value vs. Median Income")
plt.show()
```

```python
sns.scatterplot(data=df,x='AveRooms',y='MedHouseVal')
plt.title("House Value vs. Average Rooms")
plt.show()
```

```python
df.corr()['MedHouseVal'].sort_values(ascending=False) #calcuate correlation with target variable MedHouseVal
```

```python
#Correlation Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(),annot=False,cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
