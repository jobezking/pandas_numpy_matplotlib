```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]
# Create a line plot
plt.plot(x, y)
plt.title("A Simple Line Plot")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
fig, ax = plt.subplots()

x = [0, 1, 2, 3, 4]
y = [0, 1, 4, 9, 16]

# Plot on the axis
ax.plot(x,y)

# Add labels and title
ax.set_title("Quadratic Growth")
ax.set_xlabel("Input")
ax.set_ylabel("Output")

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label='Squared', color='blue', marker='o', linestyle='--')
plt.plot(x, y2, label='Linear', color='green', linewidth=2)

plt.title("Linear vs Squared Growth")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
x = [1, 2, 3, 4, 5] 
y1 = [x**2 for x in x]
y2 = [x**3 for x in x]

axs[0].plot(x, y1, color='orange')
axs[0].set_title("Squared")

axs[1].plot(x, y2, color='purple')
axs[1].set_title("Cubed")

for ax in axs:
    ax.set_xlabel("X")
    ax.set_ylabel("Y")
    ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Adding Titles
x = [1, 2, 3, 4, 5]
y = [10, 12, 15, 18, 25]

plt.plot(x, y)
plt.title("Monthly Revenue Growth") # Add a title to your chart
plt.title("Monthly Revenue Growth", fontsize=16, fontweight='bold', color='navy') #Style title
plt.show()
```

```python
#Labeling Axes
plt.plot(x, y)
plt.title("Monthly Revenue Growth")
plt.xlabel("Month") # X-axis label
plt.ylabel("Revenue (in thousands)") # Y-axis label
plt.xlabel("Month", fontsize=12, color='darkgreen')
plt.ylabel("Revenue", fontsize=12, color='darkred')
plt.show()
```

```python
#Adding a legend and annotation
x = [1, 2, 3, 4, 5]
sales = [10, 15, 20, 25, 30]
profit = [1, 2, 3, 3.5, 5]

plt.plot(x, sales, label="Sales", color="blue")
plt.plot(x, profit, label="Profit", color="green")

plt.title("Sales and Profit Over Time")
plt.xlabel("Quarter")
plt.ylabel("Value (in thousands)")

# Add annotation
plt.annotate('Profit jump', xy=(5, 5), xytext=(3, 6),
             arrowprops=dict(facecolor='red', shrink=0.05))

plt.legend()  # This adds the legend
plt.legend(loc="upper left") #changes the location
plt.grid(True)
plt.show()
```
