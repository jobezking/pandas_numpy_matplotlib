```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Built-in dataset: tips data from a restaurant
tips = sns.load_dataset("tips")
```

```python
# Histogram with kernel density estimate (KDE) 
sns.histplot(tips['total_bill'], kde=True)
plt.title("Distribution of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
plt.title("Tip vs. Total Bill")
plt.show()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex")
plt.title("Tip vs. Total Bill by Gender")
plt.show()
```

```python
sns.boxplot(data=tips, x="day", y="tip")
plt.title("Tips by Day")
plt.show()
```

```python
sns.violinplot(data=tips, x="day", y="tip", hue="sex", split=True)
plt.title("Violin Plot of Tips by Day and Gender")
plt.show()
```

```python
sns.pairplot(tips, hue="sex")
plt.suptitle("Pairwise Relationships in Tips Dataset", y=1.02)
plt.show()
```

```python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Plot histogram for total bill amount
sns.histplot(tips['total_bill'], bins=20, kde=False, color='skyblue')

plt.title('Histogram of Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='time')

plt.title('Scatter Plot of Tip vs. Total Bill')
plt.xlabel('Total Bill ($)')
plt.ylabel('Tip ($)')
plt.show()
```

```python
sns.barplot(data=tips, x='day', y='tip', errorbar=None, palette='pastel')

plt.title('Average Tip by Day')
plt.xlabel('Day of the Week')
plt.ylabel('Average Tip ($)')
plt.show()
```

```python
sns.boxplot(data=tips, x='smoker', y='tip', palette='Set2')

plt.title('Tip Distribution by Smoking Preference') 
plt.xlabel('Smoker') 
plt.ylabel('Tip ($)') 
plt.show()
```

```python
corr = tips.corr(numeric_only=True) 

sns.heatmap(corr, annot=True, cmap='coolwarm', center=0)

plt.title('Correlation Matrix of Tips Dataset')
plt.show()
```
