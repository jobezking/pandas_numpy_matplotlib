```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('sales_data.csv')
```

```python
df['price'].hist(bins=30) 
plt.title('Price Distribution')
plt.xlabel('Price') 
plt.ylabel('Frequency') 
plt.show()
```

```python
df['revenue'].hist(bins=30) 
plt.title('Revenue Distribution')
plt.xlabel('Revenue') 
plt.ylabel('Frequency') 
plt.show()
```
